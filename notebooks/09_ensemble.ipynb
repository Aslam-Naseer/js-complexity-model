```python
from datasets import load_dataset

dataset = load_dataset("aslam-naseer/js-function-complexity-processed")
train = dataset["train"]
test = dataset["test"]
val = dataset["validation"]
```

### Final models

```python
llm_agent = LLM_Agent()
nn_agent = NN_Agent()
```

```python
def nn_predict(item):
  return nn_agent.predict(item)

evaluate(nn_predict, test)
```

```python
counter = 0
def llm_predict(item):
    global counter
    score = llm_agent.predict_saved(counter)
    counter += 1
    return score

evaluate(llm_predict, test)
```

```python
NN_BIAS = 0.3
LLM_BIAS = 0.7

counter = 0
def ensemble_predict(item):
    global counter
    nn_score = nn_agent.predict(item)
    llm_score = llm_agent.predict_saved(counter)
    counter += 1
    final_score = NN_BIAS * nn_score + LLM_BIAS * llm_score
    return final_score

evaluate(ensemble_predict, test)
```
